## Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

## Loading Dataset

```python
df1 = pd.read_csv('matches.csv')
```

```python
df1.head()
```

```python
df2 = pd.read_csv('deliveries.csv')
```

```python
df2.head()
```

## Merging the dataframes

```python
#merged two dataframes using left join
df3 = pd.merge(df1,df2.rename(columns={'match_id':'id'}),on='id',  how='left')
```

```python
df3.head()
```

```python
#showing first row of the dataframe df3
df3.iloc[1,:]
```

## Visualizations

```python
#top 10 players in ipl
TopPlayers = dict(df1.player_of_match.value_counts()[:10])
x=list(TopPlayers.keys())
y=list(TopPlayers.values())

plt.figure(figsize = (12,7))
plt.bar(x, y, width= 0.9, align='center',color='orange',edgecolor='black')
plt.xlabel('Player name')
plt.ylabel('Man of the Match frequency')
plt.title('Top 10 players in IPL')
plt.show()
```

```python
#top 10 fielders in IPL
Topfielders = dict(df3.fielder.value_counts()[:10])
x=list(Topfielders.keys())
y=list(Topfielders.values())
plt.figure(figsize = (12,7))
plt.bar(x, y, width= 0.9, align='center',color='lightblue',edgecolor='black')
plt.xlabel('Player name')
plt.ylabel('Caught the ball')
plt.title('Top 10 fielders in IPL')
plt.show()
```

```python
#match won by choosing bat first or ball first result 
bat_first = len(df1[df1['win_by_runs']>0])
ball_first=len(df1[df1['win_by_wickets']>0])
labels = ['bat first', 'ball first']
sizes = [bat_first, ball_first]
colors = ['cyan', 'yellow']
plt.pie(sizes, colors=colors,autopct='%1.1f%%', shadow=True, startangle=180)
plt.legend(labels, loc="best")
plt.axis('equal')
plt.title('Match Win Result')
plt.tight_layout()
plt.show()
```

```python
#The number of matches won by each team
plt.figure(figsize = (10,10))
sns.countplot(x='winner',data=df1)
plt.title("Numbers of matches won by team ",fontsize=20)
plt.xticks(rotation=90)
plt.xlabel("Teams",fontsize=15)
plt.ylabel("Number of wins",fontsize=15)
plt.show()
```

```python
# grouping matches won by each team by season

import numpy as np
grouped = df1.groupby('season')
dic=dict()
l=list()
k=list()
for name,group in grouped:
        print(name)
        print(group['winner'].value_counts())
       
```

```python
# top 3 scorers of ipl
batsman_score=df3.groupby('batsman')['batsman_runs'].agg(['sum']).sort_values('sum',ascending=False)
print("Top 3 scorers of IPL")
batsman_score.iloc[:3,:]
```

```python
# toss results 
toss_decision = df1['toss_decision'].value_counts()
labels=['bat','field']

sizes = toss_decision.values
colors = ['cyan', 'yellow']
plt.pie(sizes, colors=colors,autopct='%1.1f%%', shadow=True, startangle=180)
plt.legend(labels, loc="best")
plt.axis('equal')
plt.title('Toss result')
plt.tight_layout()
plt.show()
```

```python
#count of different dismissal_kind
plt.figure(figsize=(12,6))
sns.countplot(x='dismissal_kind', data=df3)
plt.xticks(rotation='vertical')
plt.show()
```
